```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('P  L March 2021.csv')  # Replace 'your_dataset.csv' with the actual file path

# Drop irrelevant columns
df = df.drop(['index', 'Sku', 'Style Id', 'Catalog'], axis=1)

# Handling missing values (if any)
df = df.dropna()  # Drop rows with missing values

# Visualize the dataset

# Bar plot for category counts
category_counts = df['Category'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Counts')
plt.xticks(rotation=45)
plt.show()

# Histogram for weight distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Weight'], bins=20)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()

# Line plot for MRP Old and Final MRP Old
plt.figure(figsize=(10, 6))
plt.plot(df['MRP Old'], label='MRP Old')
plt.plot(df['Final MRP Old'], label='Final MRP Old')
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('MRP Old vs Final MRP Old')
plt.legend()
plt.show()

# Scatter plot for Amazon MRP vs Flipkart MRP
plt.figure(figsize=(50, 10))
plt.scatter(df['Amazon MRP'], df['Flipkart MRP'])
plt.xlabel('Amazon MRP')
plt.ylabel('Flipkart MRP')
plt.title('Amazon MRP vs Flipkart MRP')
plt.show()
```
